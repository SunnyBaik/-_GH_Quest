```python
import pandas as pd
```

```python
CCTV_Seoul = pd.read_csv('C:\\Users\\user\\Desktop\PythonWD\\GH\\CCTV_in_Seoul.csv', encoding='utf-8') #인코딩이 안되면 engine = 'python'으로 해보세요!
CCTV_Seoul.head()
```

```python
CCTV_Seoul.rename(columns={CCTV_Seoul.columns[0]:'구별'},inplace=True)
CCTV_Seoul.head()
```

```python
pop_Seoul = pd.read_excel('C:\\Users\\user\\Desktop\PythonWD\\GH\\population_in_Seoul.xls', 
                          header = 2, usecols = 'B, D, G, J, N', encoding = 'utf-8')
pop_Seoul.head()
```

```python
pop_Seoul.rename(columns={pop_Seoul.columns[0]:'구별',
                         pop_Seoul.columns[1]:'인구수',
                         pop_Seoul.columns[2]:'한국인',
                         pop_Seoul.columns[3]:'외국인',
                         pop_Seoul.columns[4]:'고령자'},inplace=True)
pop_Seoul.head()
```

```python
pop_Seoul.drop([0], inplace = True)
pop_Seoul.head()
```

```python
CCTV_Seoul['최근증가율'] = (CCTV_Seoul['2016년'] + CCTV_Seoul['2015년'] + CCTV_Seoul['2014년']) / CCTV_Seoul['2013년도 이전'] * 100
CCTV_Seoul.head(5)
```

```python
data_result = pd.merge(CCTV_Seoul, pop_Seoul, on = '구별')
data_result.head()
```

```python
del data_result['2013년도 이전']
del data_result['2014년']
del data_result['2015년']
del data_result['2016년']
data_result.head()
```

```python
data_result.set_index('구별', inplace=True)
data_result.head()
```

```python
data_result['외국인비율'] = (data_result['외국인']) / data_result['인구수'] * 100
data_result['고령자비율'] = (data_result['고령자']) / data_result['인구수'] * 100
data_result['CCTV비율'] = (data_result['소계']) / data_result['인구수'] * 100
```

```python
data_result.head()
```

```python
import matplotlib.pyplot as plt #matplotlib가 한글폰트를 지원 안해서 미리 처리해주는 과정
import platform                 #아마 요즘 matplotlib는 이런거 안해도 잘 될수도 있어요!

from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus']= False

if platform.system() == 'Darwin': #맥os 사용자
    rc('font', family = 'AppleGothic')
elif platform.system() == 'Windows':#윈도우 사용자
    path = 'c:/Windows/Fonts/malgun.ttf'
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
    
import numpy as np
fp1 = np.polyfit(data_result['인구수'], data_result['소계'], 1)
f1 = np.poly1d(fp1)
fx = np.linspace(100000, 700000, 100)
```

```python
data_result['오차']=np.abs(data_result['소계']-f1(data_result['인구수']))
data_result.head()
```

```python
data_result.sort_values(by = '오차', ascending = False).head()
```

```python
data_result = data_result.sort_values(by = '오차', ascending = False)
data_result.head()
```

```python
colormap = data_result['오차']
plt.figure(figsize=(10,6))
plt.scatter(data_result['인구수'], data_result['소계'], s=50, c = colormap)

for i in range(10):
    plt.text(data_result['인구수'][i]*1.02, data_result['소계'][i]*0.98, data_result.index[i], fontsize= 15)
    
plt.plot(fx, f1(fx), ls='dashed', lw=3, color ='g')
plt.xlabel('인구수')
plt.ylabel('소계')
plt.grid()
plt.colorbar()
plt.show()
```
